## FE

```python
date_columns = ['SIGN_DATE','DATEFIRSTREG','TAXREGPAY_REGDATE','TAXREG_REGDATE']
bin_columns = ['IP_flag','EGRPOINCLUDED',]
cat_columns = ['OKFS_GROUP','OKOPF_GROUP','OKOGU_GROUP','WORKERSRANGE']
real_columns = ['COUNTCOOWNERFCSM','COUNTCOOWNERROSSTAT','COUNTCOOWNEREGRUL',
        'COUNTBRANCH','COUNTBRANCHROSSTAT','COUNTBRANCHEGRUL',
       'TELEPHONECOUNT','MANAGERCOUNTINCOUNTRY','MANAGERCOUNTINREGION','MANAGERINNCOUNT']
discreate_real = ['OKATO_REGIONCODE','OKATO_FED','OKTMO_FED'] # OKTMO_CODE - just number - ignore

real_columns_F = []
for name in train.columns.values:
    if name[0] == 'F': real_columns_F.append(name)

real_columns_F_L = []
for name in train.columns.values:
    if name[0] == 'F' and name[-1] == '1': real_columns_F_L.append(name)
        
real_columns_all = real_columns + discreate_real + real_columns_F + real_columns_F_L
```

```python
import sklearn.preprocessing as prp
```

```python
def make_date_features(data, date_columns):
    for name in date_columns:
    # the time
        data[name] = pd.to_datetime(train[name], format = '%d%b%Y:%H:%M:%S', errors = 'coerce')
        data[name + '_day'] = data[name].dt.day
        data[name + '_month'] = data[name].dt.month
        data[name + '_dayofweek'] = data[name].dt.dayofweek
        data[name + '_dayofyear'] = data[name].dt.dayofyear
    # isnull
        data['null_in_'+ name] = data[name].isnull().astype(int)
    # fill the nan
    for name in date_columns:
        data[name].fillna(data[name].mean(),inplace = True)

    # diff
    data['TAXREG_REGDATE_DATEFIRSTREG'] = (data['TAXREG_REGDATE'] - data['DATEFIRSTREG']).astype("timedelta64[D]")
    data['SIGN_DATE_DATEFIRSTREG'] = (data['SIGN_DATE'] - data['DATEFIRSTREG']).astype("timedelta64[D]")
    
    return data

def make_bin_features(data, bin_columns):
    # isnull
    for name in bin_columns:
        data['null_in_'+ name] = data[name].isnull().astype(int)
        # fill with mean()： binary to average
        data[name].fillna(data[name].mean(),inplace = True)
    return data

def make_cat_features(data, cat_columns):
    # fill with random str:
    for name in cat_columns:
        data[name].fillna('NotANumber',inplace = True)
    # OHE 
    data = pd.get_dummies(data, columns = cat_columns)
    return data

def make_real_features(data, real_columns):
    # fill with ave:
    for name in real_columns:
        data[name].fillna(data[name].mean(),inplace = True)
    # set barriers
    real_fea_map = {'COUNTCOOWNERFCSM':8 ,
               'COUNTCOOWNERROSSTAT':5,
               'COUNTCOOWNEREGRUL': 12,
               'COUNTBRANCH':7,
               'COUNTBRANCHROSSTAT':50,
               'COUNTBRANCHEGRUL':60,
               'MANAGERCOUNTINCOUNTRY':75}
    for name in real_columns:
        if name in real_fea_map:
            data[name + 'barrier'] = real_fea_map[name]
            
    # aggregation
    data['real_mean'] = data[real_columns].mean(axis=1)
    data['real_std'] = data[real_columns].std(axis=1)
    data['real_max'] = data[real_columns].max(axis=1) 
    data['real_min'] = data[real_columns].min(axis=1)
    
    # isnull
    data['null_in_'+ name] = data[name].isnull().astype(int)
    
    return data

def make_norm(data, cols):
    for name in cols:
        data[name] = prp.StandardScaler().fit_transform(data[[name]])
        
    return data

def make_features(data):
    data = make_date_features(data, date_columns)
    # data = make_bin_features(data, bin_columns)
    data = make_cat_features(data, cat_columns)
    # data = make_real_features(data, real_columns)
    # data = make_real_features(data, discreate_real)
    # data = make_real_features(data, real_columns_F)
    # data = make_real_features(data, real_columns_F_L)
    # data = make_norm(data, real_columns_all)
    
    return data
```

```python
train_fe = make_features(train)
test_fe = make_features(test)
print(f'the shape of FE train: {train_fe.shape}')
train_fe.head()
```

## baseline - logistic regression

```python
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LogisticRegression

label_col = 'TARGET'
features_cols = train.drop('TARGET', axis='columns').select_dtypes(include=[np.number]).columns.tolist()
```

```python
logit = LogisticRegression(max_iter=10000)

cv_baseline_score = cross_val_score(logit, train[features_cols].fillna(0), train[label_col], cv=5, scoring='roc_auc').mean()
print(f'roc_auc_score: {cv_baseline_score}')

logit.fit(train[features_cols].fillna(0), train[label_col])
```

```python
test['TARGET'] = logit.predict_proba(test[features_cols].fillna(0))[:, 1]

test[['id_contract', 'TARGET']].to_csv('submit_file.csv', sep=';', index=False)
test[['id_contract', 'TARGET']].head()
```

## baseline random forest

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import cross_val_score

label_col = 'TARGET'
features_cols = train.drop(['TARGET','id_contract','id_client'], axis='columns').select_dtypes(include=[np.number]).columns.tolist()
```

```python
# random run
rf = RandomForestClassifier(n_estimators = 100, max_depth = 10, n_jobs = -1, random_state = 1)

cv_baseline_score = cross_val_score(rf, train_fe[features_cols].fillna(0), train_fe[label_col], cv=5, scoring='roc_auc').mean()
print(f'roc_auc_score: {cv_baseline_score}')

rf.fit(train_fe[features_cols].fillna(0), train_fe[label_col])
```

```python
# tuning: n_estimators = 80, max_depth =  13, min_samples_split  = 70 
param_test1 = {'n_estimators':range(10,100,10)}
gsearch1 = GridSearchCV(estimator = RandomForestClassifier(min_samples_split=100,
                                  min_samples_leaf=20,max_depth=8,max_features='sqrt' ,random_state=10), 
                       param_grid = param_test1, scoring='roc_auc',cv=5)
gsearch1.fit(train_fe[features_cols].fillna(0), train_fe[label_col])
gsearch1.best_params_, gsearch1.best_score_
```

```python
param_test2 = {'max_depth':range(3,14,2), 'min_samples_split':range(50,201,20)}
gsearch2 = GridSearchCV(estimator = RandomForestClassifier(n_estimators= 70, 
                                  min_samples_leaf=20,max_features='sqrt' ,oob_score=True, random_state=10),
   param_grid = param_test2, scoring='roc_auc', cv=5)
gsearch2.fit(train_fe[features_cols].fillna(0), train_fe[label_col])
gsearch2.best_params_, gsearch2.best_score_
```

```python
# use tuned paras
rf = RandomForestClassifier(n_estimators = 70, max_depth = 13, min_samples_split = 50, n_jobs = -1)

cv_baseline_score = cross_val_score(rf, train_fe[features_cols].fillna(0), train_fe[label_col], cv=5, scoring='roc_auc').mean()
print(f'roc_auc_score: {cv_baseline_score}')

rf.fit(train_fe[features_cols].fillna(0), train_fe[label_col])
```

```python
test['TARGET'] = rf.predict_proba(test_fe[features_cols].fillna(0))[:, 1]

test[['id_contract', 'TARGET']].to_csv('submit_file_rf_fe_3.csv', sep=';', index=False)
test[['id_contract', 'TARGET']].head()
```
